# **1. BAR CHART**

**The above chart is implemented using the dataFrame1, which includes the daily number of customer visits to each outlet. The main purpose of this bar chart is to visualize the high, medium and low volume of outlets. The visualization is implemented with different colors for better identification of high, medium and low volumes. This is done using the if statement where the range of each volume has been mentioned in the python code. The different range of values will be assigned a different color.**

**Each of the bars represent the sum of each outlet.**

```python
colours = []
for name in dataFrame1.columns:
    total_visits = dataFrame1[name].sum()
    if total_visits > 400000: #range set for high volume of outlet visits
        colour = '#ff0011'
    elif total_visits > 200000: #range set for medium volume of outlet visits
        colour = '#f005c9'
    elif total_visits > 8914: #range set for low volume of outlet visits
        colour = '#0520f0'
    else:
        colour = '#10630d' #very low volume of outlet visits
    colours.append(colour)

plt.figure(figsize=(25, 8))#figure size
x_pos = np.arange(len(dataFrame1.columns))
plt.bar(x_pos, dataFrame1.sum(), align='center', color=colours)
plt.xticks(x_pos, dataFrame1.columns,rotation='vertical',weight = 'bold', fontsize=17)
plt.grid(color='#b6c2d6', linestyle='--', linewidth=2.3, axis='y', alpha=0.5)
plt.xlabel('ChrisCos Outlets', fontsize=21,weight = 'bold')
plt.ylabel('Daily Number of Customer Visits', fontsize=21,weight = 'bold')
plt.title('Sum of Daily Number of Customer Visits at ChrisCos Outlets', fontsize=23,weight = 'bold')
plt.yticks(weight = 'bold', fontsize=15) 
plt.show()
```

**It can be seen in the above bar chart, the bar chart has been categorized using 4 different colors. The color red represents high number of daily customer visits, the color pink represent medium number of daily customer visits, the color blue represents low number of daily customer visits and finally the color green represents a very low volume of daily customer visits. In order to set different colors, the range for each volume has been set, where visits greater than 400000 is categorized as high volume and is the given the color red, the visits greater than 200000 is categorized as medium volume and is given the color pink, the visits greater than 8914 is categorized is low volume and is given the color blue and lastly, the visits less than that is given the color green and is categorized as very low volume of outlet visits.The dataset is sorted which is why the bar chart is shown in an increasing order starting from high number of customer visit to a very low number of customer visit.**

**From the above Bar Chart of Sum of Daily Number of Customer Visits at ChrisCos Outlets, it can be identified that the 3 red bars are the outlets with a high volume of visits and they are: RAN, RFY AND DMN. The 7 pink bars represent the outlets EYS,DSA,EEC,BMF,BSQ,CYK,CGV and these 7 outlets have medium number of of daily Customer Visits and the rest of the outlets have a low number of daily customer visits and is shown in blue and green bars.**

# 2. Line Plot with Monthly Averages for Outlets with a High and Medium Volume of Visits 

**I have used the Line Plot to show changes in outlet visits over a year with monthly averages for viewer to understand and know in one glimpse about the average volume of outlet visits over a year. It is difficult to visualize the average of a month just from viewing the line plot done based on daily visits as the number of data points in the plot is too much to understand or find any trends or patters. Hence this is why I have implemented it with monthly averages for an easier visualization of trends and patters of the outlets with a high and medium volume of visits.** 

```python
sample_rate = 'M'  # monthly averages
averaged_data = dataFrame1.resample(sample_rate).mean()#averaged dataset

#selecting the outlets with a high and medium volume of visits
selected = ['RAN', 'RFY', 'DMN','EYS','DSA','EEC','BMF','BSQ','CYK','CGV']

plt.figure(figsize=(11, 10))#figure size
plt.plot(dataFrame1[selected], linewidth=0.5)#line plot
plt.gca().set_prop_cycle(None)
plt.plot(averaged_data[selected], linewidth=4)#line plot of average values
plt.xlabel('Date', fontsize=18,weight = 'bold')
plt.ylabel('Number of Customer Visits', fontsize=18,weight = 'bold')
plt.title('High and Medium Volume of Customer Visits to Outlets with Monthly Averages', fontsize=20,weight = 'bold')
plt.legend(selected, loc=2, fontsize="11.9")
plt.yticks(weight = 'bold', fontsize=13)
plt.xticks(weight = 'bold', fontsize=13)
plt.ylim(ymin=450)#setting a minimum value of the y axis
plt.show()
```

**It is found earlier, that the outlets 'RAN', 'RFY', 'DMN' are the outlets with a high volume of customer visits and the outlets 'EYS','DSA','EEC','BMF','BSQ','CYK','CGV' are the outlets with a medium volume of customer visits.The above plot is a Line Plot, with monthly averages of customer visits to high and medium volume of outlets from the date 2021-01 to 2022-01.**

**In the plot each of the outlets with a high and medium volume of customer visits are displayed with different colors for identification. The thick lines in the line plot are the lines that show the average number of customer visits to the outlets. These lines also act as a trendline which shows whether if the volume of customer visits have increased or decreased with time.**

**From the line plot, it can be said that from the high volume outlets RAN and DMN, have a gradual increasing trend from 2021-09, however RFY has an overall fluctuating trend with a fall after 2021-12.**

# 5. HEAT MAP for ChrisCo's Summary Data

**The above heat map is implemented using the summary data in order to find the correlation between each of the columns that includes marketing, customer data, outlet size, outlet staff and overheads. It is important to find out about the correlations for better decision making. For example- if it can be found that there is a positive correlation between customer visits and marketing, then this will help ChrisCo in taking future decisions.
```python
#heat map of combined summary data
plt.figure(figsize=(12, 9))#figure size
corr = ChrisCo_Summary_Data.corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.cubehelix_palette(as_cmap=True) , square=True, annot=True,annot_kws={"size": 14}, linewidth=.6)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, horizontalalignment='right',weight='bold',fontsize=12)
ax.set_yticklabels(ax.get_xticklabels(), rotation=30, horizontalalignment='right',weight='bold',fontsize=12)
ax.set_title('Heat Map of ChrisCos Summary Data',weight='bold',fontsize=16)
plt.show()
```

# 6. Box Plot for ChrisCo's Summary Data

**Here I have implemented a box plot for summary data. I have selected all the columns for this box plot except the columns with customer visits, as I have previously already found information regarding this and I wanted to focus only on these 4 columns. I have used different y limit for each of the box sub plots, in order to get a clearer view of all the box plot. I have used a dotted line to show the value of median for better visualization and understanding. In this way the viewer does not have to see the y axis to find the median value and can rather view it directly from the dotted line.**

**The purpose of implementing the box plot is to find the maximum, median and minimum values in addition to the outliers each columns may have. The outliers is important to find out as it is an extreme value and is an important thing to consider when making decisions as it may indicate seasonality or errors.**

```python
#Box plot for ChrisCo's Summary Data
#I have selected other columns except Customer Data as I want to only find information regarding marketing, overheads, 
#outlet staff and outlet size
selected = ['Marketing', 'Overhead', 'Outlet_Staff', 'Outlet_Size']
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))#subplots of 2 rows and 2 columns

#Different y limits for each of the 4 subplots
y_range = {'Marketing': (0, 77000), 'Overhead': (0, 100000),
            'Outlet_Staff': (0, 50), 'Outlet_Size': (0, 5100)}
# Creating the boxplot for the present column
for m, column in enumerate(selected):
    #row and column index
    row = m // 2
    col = m % 2
    
    # Creating the boxplot for the present column
    ax[row, col].boxplot(ChrisCo_Summary_Data[column])
    ax[row, col].set_ylabel('Value', fontsize=12,weight='bold')
    ax[row, col].set_ylim(y_range[column])  
    ax[row, col].set_title(column, fontsize=14,weight='bold')
    ax[row, col].grid(True)#gridlines
    #Code for displaying the median value.
    median = np.median(ChrisCo_Summary_Data[column])   
    ax[row, col].plot(1, median, marker='s', markersize=5, color='b', label='Median')
    ax[row, col].plot([0.8, 1.2], [median, median], linestyle='--', color='b', linewidth=1.7)
    ax[row, col].text(1.22, median, f'{median:.2f}', fontsize=12,weight='bold')
    ax[row, col].legend(loc='upper right')    
        
plt.suptitle('Box Plot for ChrisCos Summary Data', fontsize=17,weight='bold')
plt.tight_layout()
plt.show()
```

**The 4 sub box plots shows the median, maximum, minimum values and the outliers. Other than overheads, all other subplots contains outliers which may be due to seasonality or for errors. It can also be seen that other than overheads, all other bar sub plots has a narrow data distribution.**

# 7. Bar Subplots for ChrisCo's Summary Data

**Here, I have implemented subplots of bar chart using ChrisCo's Summary Data. Same as last plot, I have selected other columns except the column with customer visit as I only want to focus on information regarding the 4 columns which are marketing, overhead, outlet staff and outlet size. The purpose of this bar plot is to find out which outlets have the highest and lowest size, cost and staff. This information was not possible to find with a box plot, as it does not take the outlets into consideration.**

```python
#Bar subplots for ChrisCo's Summary Data
plt.figure(figsize=(25, 25))#figure size
counter = 1
selected=['Marketing','Overhead','Outlet_Staff','Outlet_Size']#selected columns

x_pos = np.arange(len(ChrisCo_Summary_Data.index))
for attributes in ChrisCo_Summary_Data[selected]:
    
    sub = plt.subplot(4,1, counter)#subplot of 4 rows and 1 column
    sub.bar(x_pos, ChrisCo_Summary_Data[attributes].sort_values(ascending=False), align='center',color='#4dc481')
    sub.set_xticks([])
    sub.set_xticks(x_pos)
    sub.tick_params(axis='both', labelsize=14)  # Set ytick font size and weight
    sub.set_xticklabels(ChrisCo_Summary_Data.index,rotation=56,fontweight='bold', fontsize=19)
    sub.set_xlabel('ChrisCos Outlets', fontsize=17,fontweight='bold')
    sub.set_ylabel(attributes, fontsize=24,fontweight='bold')
    counter += 1

plt.tight_layout()
plt.show()
```

**From the above plot, it can be identified that the 10 outlets identified previously as high and medium volume of outlets are the outlets that have the highest marketing cost. In addition, these 10 outlets have the highest outlet size and outlet staff. On the otherhand, the overhead cost for most of the outlet is over 20000.**

# 8. Autocorrelation Plot of Outlets with High Volume of Customer Visits.

**For this plot, I have implemented an autocorrelation plot of outlets with a high volume of customer visits. The purpose of this plot is to identify if there is any seasonality.**

```python
row = 0
col = 0
selected=['RAN','RFY','DMN']#selected high volume outlets

fig, axes = plt.subplots(figsize=(10, 10), nrows=3, ncols=1)#subplot of 3 rows 1 column
fig.suptitle('Autocorrelation plots', fontsize=20, position=(0.55, 0.97),weight='bold')
for name in dataFrame1[selected]:
    sub = pd.plotting.autocorrelation_plot(dataFrame1[name], ax=axes[row])
    sub.set_title('Outlet Name: ' + name, fontsize=16,weight='bold')
    sub.set_ylabel("Autocorrelation", fontsize=14,fontweight='bold')
    sub.set_xlabel("Lag", fontsize=14,fontweight='bold')
    row += 1
plt.subplots_adjust(wspace=0.5, hspace=0.4)#space between subplots
plt.show()
```

**From the above plot, it can be seen that there is seasonality in all 3 of the subplots.**
